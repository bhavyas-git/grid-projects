```python
from src.core.dataset import get_dataloaders
from src.models.mobilenetV3 import get_mobilenet_v3
from src.core.train import train_model
from src.core.losses import FocalLoss
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

train_loader, val_loader = get_dataloaders("../data/raw/train", "../data/raw/test")

model = get_mobilenet_v3(num_classes=7).to(device)

criterion = FocalLoss()
optimizer = optim.AdamW(model.parameters(), lr=3e-4)

scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)

train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, device)
```
